## SimRanK

El objetivo de esta libreta es utilizar el algoritmo de SimRank para lo cual se empleará la biblioteca [simrank_similarity](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.similarity.simrank_similarity.html) del módulo de networkx.

### Ejemplo sencillo de SimRank

```python
"""Carga de bibliotecas"""
from numpy import array
import networkx as nx
```

```python
G = nx.cycle_graph(4)
nx.draw(G, with_labels = True) #Visualizamos el grafo
```

```python
sim = nx.simrank_similarity(G,target = 2)
```

```python
sim
```
