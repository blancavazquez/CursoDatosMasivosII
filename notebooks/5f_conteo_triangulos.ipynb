## Conteo de triángulos

El objetivo de esta libreta es emplear el algoritmo de conteo de triángulos usando las funciones de agrupamiento de la biblioteca de [Networkx](https://networkx.org/documentation/stable/reference/algorithms/clustering.html)

```python
"""Carga de bibliotecas"""
from numpy import array
import networkx as nx
```

```python
G = nx.complete_graph(3)
nx.draw(G, with_labels = True) #Visualizamos el grafo
print(list(nx.triangles(G, (0, 1)).values()))
number_of_triangles = sum(nx.triangles(G).values()) / 3
print("Número de triángulos en el grafo: ", number_of_triangles)
```
